## Loading Libraries

```
from surprise import Dataset
from auto_surprise.engine import Engine

import random
import numpy
```

## Loading Data

```
# Load the dataset
data = Dataset.load_builtin('ml-100k')
```

## Using RMSE as a metric

```
# Setting seed for reproducibility
random.seed(123)
numpy.random.seed(123)

# Intitialize auto surprise engine
engine_rmse = Engine(verbose=True, random_state=numpy.random.RandomState(123))

# Start the trainer
best_algo_rmse, best_params_rmse, best_score_rmse, tasks_rmse = engine_rmse.train(
    data=data, 
    target_metric='test_rmse', 
    cpu_time_limit=60 * 60, 
    max_evals=100
)
```

## Using MAE as metric

```
# Setting seed for reproducibility
random.seed(123)
numpy.random.seed(123)

# Intitialize auto surprise engine
engine_mae = Engine(verbose=True, random_state=numpy.random.RandomState(123))

# Start the trainer
best_algo_mae, best_params_mae, best_score_mae, tasks_mae = engine_mae.train(
    data=data, 
    target_metric='test_mae', 
    cpu_time_limit=60 * 60, 
    max_evals=100
)
```

```
def get_loss_dict(task_dict):

  loss_dict = {}
  
  for task in task_dict.keys():
    cur_loss = task_dict[task]['loss']

    if not cur_loss:
      cur_loss = 0

    loss_dict[task] = cur_loss
  
  return loss_dict



task_rmse_loss_dict = get_loss_dict(tasks_rmse)
task_mae_loss_dict = get_loss_dict(tasks_mae)
```

## Using Hyperopt with AutoRecSys

## Using RMSE as metric

```
# Example for setting the HPO algorithm to adaptive TPE
import hyperopt

# Setting seed for reproducibility
random.seed(123)
numpy.random.seed(123)

# Intitialize auto surprise engine
engine_hopt_rmse = Engine(verbose=True, random_state=numpy.random.RandomState(123))

# Start the trainer
best_algo_hopt_rmse, best_params_hopt_rmse, best_score_hopt_rmse, tasks_hopt_rmse = engine_hopt_rmse.train(
    data=data, 
    target_metric='test_rmse',
    cpu_time_limit=60 * 60,
    max_evals=100,
    hpo_algo=hyperopt.tpe.suggest
)
```

```
print(best_algo_hopt_rmse) 
print(best_params_hopt_rmse) 
print(best_score_hopt_rmse) 
print(tasks_hopt_rmse)
```
