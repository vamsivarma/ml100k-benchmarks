```
import numpy as np
import matplotlib.pyplot as plt
```

```
def plot_hbar(plt_dict, xlabel, plt_title, b_color):
  # Fixing random state for reproducibility
  np.random.seed(19680801)


  plt.rcdefaults()
  fig, ax = plt.subplots()

  # Example data
  y_pos = np.arange(len(plt_dict))

  ax.barh(y_pos, plt_dict.values(), color=b_color)
  ax.set_yticks(y_pos)
  ax.set_yticklabels(plt_dict.keys())
  ax.invert_yaxis()  # labels read top-to-bottom
  ax.set_xlabel(xlabel)
  ax.set_title(plt_title)
  plt.show()

```
