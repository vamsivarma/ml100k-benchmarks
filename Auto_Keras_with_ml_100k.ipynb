```python
import pandas
import numpy

# example of auto-sklearn for the ml-100k regression dataset
from pandas import read_csv
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error

# use autokeras to find a model for the ml-100k dataset
from numpy import asarray
from autokeras import StructuredDataRegressor
```

```python
# load dataset
url = 'ratingss.csv'
dataframe = read_csv(url)
```

```python
# load dataset

# split into input and output elements
data = dataframe.values
data = data.astype('float32')
X, y = data[:, :-1], data[:, -1]
# split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=1)
```

## Evaluating MAE

```python
# define the search
search = StructuredDataRegressor(max_trials=10, loss='mean_absolute_error')

# perform the search
search.fit(x=X_train, y=y_train, verbose=1, epochs=100)

# evaluate the model
mae, _ = search.evaluate(X_test, y_test, verbose=0)
print('MAE: %.3f' % mae)
```

```python
mae
```

## Evaluating RMSE

```python
# define the search
search_rmse = StructuredDataRegressor(max_trials=10, loss='mean_squared_error')

# perform the search
search_rmse.fit(x=X_train, y=y_train, verbose=1, epochs=100)

# evaluate the model
rmse, _ = search_rmse.evaluate(X_test, y_test, verbose=0)
print('RMSE: %.3f' % rmse)
```

```python
rmse
```

## Save the best model for future reference

```python
# get the best performing model
model = search.export_model()
# summarize the loaded model
model.summary()
# save the best performing model to file
model.save('model_ml100k_mae.tf')
```
